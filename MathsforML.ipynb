```python
#Dot pdt of 2 vectors

a = np.array([2,3])
b= np.array([4,4])

a_dot_b = np.dot(a,b)

a_dot_b
```

```python
#Dot pdt of 2 vectors

a = np.array([2,3])
b= np.array([4,4])
c= np.array([40,20,25])
d= np.array([30,5,20])
a_dot_b = np.dot(a,b)
c_dot_d = np.dot(c,d)
c_dot_d
```

```python
#cross pdt

a = np.array([2,3])
b= np.array([4,4])

a_cross_b= np.cross(a,b)

a_cross_b
```

```python
c= np.array([40,20,25])
d= np.array([30,5,20])

c_cross_d = np.cross(c,d)

c_cross_d
```

```python
#projection of one vector on another

a = np.array([2,5])
v = np.array([8, -5])

magnitude_of_v= np.sqrt(np.sum(v**2))

projection_of_a_on_V = (np.dot(a,v)/magnitude_of_v**2)*v

projection_of_a_on_V
```
